```python
# concat 

data_to_add = {'Survived':[1,0], 'Pclass':[1,3], 'Name':['Miss. Rose Dewitt Bukater', 'Mr. Jack Dawson'], 'Sex':['female', 'male'], 'Age':[17, 20], 'SibSp':[2,0], 'Parch':[0,0]}
addendum = pd.DataFrame(columns = titanic.columns, data=data_to_add)
addendum
```

```python
# concatanation

pd.concat([titanic, addendum])
```

```python
# merging 

data_to_add = {'Pclass':[1, 2, 3], 'Total cabin amount':[371, 105, 84]}
extra_info = pd.DataFrame(columns = ['Pclass', 'Total cabin amount'], data = data_to_add)
extra_info
```

```python
titanic.merge(extra_info, how='left', on='Pclass')
```
